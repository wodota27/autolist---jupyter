```python
#MULTIPLE PAGES
headers = {
    'Accept': '*/*',
    'Accept-Language': 'en-US,en;q=0.9',
    'Connection': 'keep-alive',
    'Origin': 'https://www.autolist.com',
    'Referer': 'https://www.autolist.com/',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36',
    'sec-ch-ua': '"Chromium";v="110", "Not A(Brand";v="24", "Google Chrome";v="110"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
}

model = []
city = []
price = []
mileage = []
year = []
dealername = []


for i in range(1,4):
    params = {
        'city': 'Abu Dhabi',
        'latitude': '24.4638',
        'limit': '20',
        'longitude': '54.363',
        'make': 'Ford',
        'page': str(i),
        'radius': '50',
        'state': 'AZ',
    }

    response = requests.get('https://production-proxy.autolist.com/api/cwv/seo/listings', params=params, headers=headers)
    
    results_json  = response.json()
    result_items = response.json()['search_results']
    
    
    for result in result_items:
        model.append(result['model'])
        city.append(result['city'] + result_items[0]['state'])
        price.append(result['price'])
        mileage.append(result['mileage'])
        year.append(result['year'])
        dealername.append(result['dealer_name'])
```

```python
ford_df = pd.DataFrame({'Model':model, 'City':city, 'Price':price, 'Miles':mileage, 'Year':year, 'Dealer':dealername})
```
